```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("CarPrice.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset['CarName']
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

##Car Name

```python
dataset['CarName'] = dataset['CarName'].str.replace("-"," ")
```

```python
car_name_en = LabelEncoder()
dataset['CarName'] = car_name_en.fit_transform(dataset['CarName'])
```

```python
dataset.head()
```

##FuelType

```python
fueltype_en = LabelEncoder()
dataset['fueltype'] = fueltype_en.fit_transform(dataset['fueltype'])
```

## Aspiration

```python
aspiration_en = LabelEncoder()
dataset['aspiration'] = aspiration_en.fit_transform(dataset['aspiration'])
```

## DoorNumber

```python
doornumber_en = LabelEncoder()
dataset['doornumber'] = doornumber_en.fit_transform(dataset['doornumber'])
```

## CarBody

```python
carbody_en = LabelEncoder()
dataset['carbody'] = carbody_en.fit_transform(dataset['carbody'])
```

##DriveWheel

```python
drivewheel_en = LabelEncoder()
dataset['drivewheel'] = drivewheel_en.fit_transform(dataset['drivewheel'])
```

##EngineLocation

```python
enginelocation_en = LabelEncoder()
dataset['enginelocation'] = enginelocation_en.fit_transform(dataset['enginelocation'])
```

## Engine Type

```python
enginetype_en = LabelEncoder()
dataset['enginetype'] = enginetype_en.fit_transform(dataset['enginetype'])
```

## Cycliner Number

```python
cylinernumber_en = LabelEncoder()
dataset['cylindernumber'] = cylinernumber_en.fit_transform(dataset['cylindernumber'])
```

## Fuel System

```python
fuelsystem_en = LabelEncoder()
dataset['fuelsystem'] = fuelsystem_en.fit_transform(dataset['fuelsystem'])
```

```python
sns.boxplot(x="wheelbase", data=dataset)
plt.show()
```

## StandarScaler

```python
ss = StandardScaler()
newData = pd.DataFrame(ss.fit_transform(dataset), columns=dataset.columns)
newData.head(5)
```

```python
newData.to_csv('car_price_for_machinelearingCleaned.csv')
```
